```python
import os
import docx
import PyPDF2
import re
import pandas as pd
import nltk
from collections import Counter


# Helper functions for text extraction
def extract_text_from_docx(file_path):
    """Extract text from DOCX file."""
    doc = docx.Document(file_path)
    return '\n'.join([para.text for para in doc.paragraphs])

def extract_text_from_pdf(file_path):
    """Extract text from PDF file."""
    text = ''
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text

# Clean the extracted text
def clean_text(text):
    """Clean the text by lowercasing, removing numbers, punctuation, and extra spaces."""
    text = text.lower()
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Tokenize and remove stopwords
def tokenize_text(text):
    """Tokenize text and remove stopwords."""
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [w for w in word_tokens if w not in stop_words]
    return filtered_text

# Grouping common keywords
skills_keywords = ['react', 'javascript', 'python', 'java', 'sql', 'html', 'css', 'node', 'docker']
education_keywords = ['bachelor', 'master', 'phd', 'degree', 'university', 'college']
experience_keywords = ['developer', 'engineer', 'intern', 'manager', 'lead']

def categorize_words(tokens, skills_keywords, education_keywords, experience_keywords):
    """Categorize words into predefined categories like skills, education, and experience."""
    skills = [word for word in tokens if word in skills_keywords]
    education = [word for word in tokens if word in education_keywords]
    experience = [word for word in tokens if word in experience_keywords]
    
    return skills, education, experience

# Function to extract text from resumes (both DOCX and PDF)
def extract_resume_text(file_path):
    """Extract text based on file extension."""
    if file_path.endswith('.docx'):
        return extract_text_from_docx(file_path)
    elif file_path.endswith('.pdf'):
        return extract_text_from_pdf(file_path)
    else:
        return None

# Analyze resumes and create a DataFrame
def analyze_and_group_resumes(directory, skills_keywords, education_keywords, experience_keywords):
    """Extract, clean, and group the most common words from resumes into a DataFrame."""
    resume_data = []

    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".docx") or file.endswith(".pdf"):  # Process DOCX and PDF files
                file_path = os.path.join(root, file)
                try:
                    # Extract text
                    text = extract_resume_text(file_path)
                    if text:
                        # Clean and tokenize text
                        clean_resume = clean_text(text)
                        tokens = tokenize_text(clean_resume)
                        
                        # Categorize words
                        skills, education, experience = categorize_words(tokens, skills_keywords, education_keywords, experience_keywords)
                        
                        # Store results in the DataFrame
                        resume_data.append({
                            'file_name': file,
                            'resume_text': text,
                            'skills': ', '.join(skills),
                            'education': ', '.join(education),
                            'experience': ', '.join(experience)
                        })
                except Exception as e:
                    print(f"Error processing {file}: {e}")

    # Create a DataFrame
    resume_df = pd.DataFrame(resume_data)
    return resume_df

# Example usage: Specify the directory where your resumes (DOCX and PDF) are located
resumes_directory = "C:\\Users\\Vikra\\Downloads\\Resumes\\Resumes" # Adjust this path

# Analyze resumes and create DataFrame
resumes_df = analyze_and_group_resumes(resumes_directory, skills_keywords, education_keywords, experience_keywords)

# Show the first few rows of the DataFrame
print(resumes_df.head())

# Save the DataFrame to a CSV file for further analysis
resumes_df.to_csv('resume_analysis.csv', index=False)
```
